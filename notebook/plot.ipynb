```python
import json

result_file = 'train_result.json'

with open(result_file) as f:
    results_dict = json.loads(f.read())
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(results_dict['tut3-model.pt']['Train Epoch Losses'], label='Train Loss')
plt.plot(results_dict['tut3-model.pt']['Validation Epoch Losses'], label='Valid Loss')
plt.xlabel('Epochs')
plt.xticks(np.arange(20))
plt.ylabel('Cross Entropy Loss')
plt.legend()
plt.show()
```
